```python
import pandas as pd
import json
## make your output file name a .jsonl
def df_to_jsonl(df, system_cols, user_cols, assistant_cols, output_file):
    with open(output_file, 'w') as f:
        for _, row in df.iterrows():
            system_content = " ".join(str(row[col]) for col in system_cols)
            user_content = " ".join(str(row[col]) for col in user_cols)
            assistant_content = " ".join(str(row[col]) for col in assistant_cols)

            data = {
                "messages": [
                    {"role": "system", "content": system_content},
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": assistant_content}
                ]
            }
            f.write(json.dumps(data) + '\n')
```

```python
## make your output file name a .jsonl
def csv_to_jsonl(input_csv, system_cols, user_cols, assistant_cols, output_file):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(input_csv)

    # Convert DataFrame to JSONL
    with open(output_file, 'w') as f:
        for _, row in df.iterrows():
            system_content = " ".join(str(row[col]) for col in system_cols)
            user_content = " ".join(str(row[col]) for col in user_cols)
            assistant_content = " ".join(str(row[col]) for col in assistant_cols)

            data = {
                "messages": [
                    {"role": "system", "content": system_content},
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": assistant_content}
                ]
            }
            f.write(json.dumps(data) + '\n')
```

```python
def text_to_jsonl(input_txt, system_cols, user_cols, assistant_cols, output_file):
    # Load the text file into a DataFrame
    df = pd.read_csv(input_txt, delimiter=' ', header=None)

    # Convert DataFrame to JSONL
    with open(output_file, 'w') as f:
        for _, row in df.iterrows():
            system_content = " ".join(str(row[col]) for col in system_cols)
            user_content = " ".join(str(row[col]) for col in user_cols)
            assistant_content = " ".join(str(row[col]) for col in assistant_cols)

            data = {
                "messages": [
                    {"role": "system", "content": system_content},
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": assistant_content}
                ]
            }
            f.write(json.dumps(data) + '\n')
```
