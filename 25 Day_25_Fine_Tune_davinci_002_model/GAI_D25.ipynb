```python
from openai import OpenAI
client = OpenAI(
    api_key = ''
)

client.files.create(
  file=open("data_file.jsonl", "rb"),
  purpose="fine-tune"
)
```

```python
client.fine_tuning.jobs.create(
  training_file="file-lL3pN8Yrj9a9zVGsqUpQIgIr",
  model="davinci-002"
)
```

```python
client.fine_tuning.jobs.retrieve("ftjob-3QLwhQOpvfj5yGzvMqL5IcAi")
```

```python
completion = client.completions.create(
  model="ft:davinci-002:personal::A6x96I0e",
  prompt = "What will I learn in this internship?",
  max_tokens=100
  )

print(completion.choices[0].text)
```
