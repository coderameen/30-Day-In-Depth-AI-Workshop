```python
# Import necessary libraries
import openai

# Set up OpenAI API key (replace with your own API key)
openai.api_key = ''

# Function to classify text using GPT-3 zero-shot
def gpt3_zero_shot_classification(prompt):
    response = openai.Completion.create(
        model="gpt-3.5-turbo-instruct",
        prompt=prompt,
        max_tokens=50,
        temperature=0.3
    )
    return response['choices'][0]['text'].strip()

# Example of zero-shot classification
text_to_classify = "The food at the restaurant was amazing!"
prompt = f"Classify the sentiment of the following review as positive, negative, or neutral:\n\n{text_to_classify}"
output = gpt3_zero_shot_classification(prompt)
print("GPT-3's Classification:", output)
```

```python
text_to_classify = "Bitcoin is a popular cryptocurrency."
prompt = f"Classify the topic of the following text as 'Finance', 'Technology', or 'Entertainment':\n\n{text_to_classify}"
output = gpt3_zero_shot_classification(prompt)
print("GPT-3's Classification:", output)
```

```python
text_to_classify = "Can you help me reset my password?"
prompt = f"Classify the intent of the following message as 'Support Request', 'General Inquiry', or 'Feedback':\n\n{text_to_classify}"
output = gpt3_zero_shot_classification(prompt)
print("GPT-3's Classification:", output)
```
